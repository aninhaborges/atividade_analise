```python
"Equipe de Desenvolvimento - Tarefa de Programação - P1"
#1.1 Entrada e saída de dados
def pede_num()->list:
    "Pede dois números e cria uma lista com eles"
    i=1
    lista=[]
    while i<=2:
        num=input('número:')
        lista.append(float(num))
        i+=1
    return lista

def media(lista:list)->float:
    'Calcula a média dos dois números da lista'
    soma=0
    for n in lista:
        soma+=n
    return soma/2

def main()->str:
    "Função principal"
    lista=pede_num()
    m=media(lista)
    print('A média é ',m)

#1.2 Estruturas de controle
def pergunta()->str:
    "Pergunta o usuário qual a sua equipe"
    resposta= input('Qual a sua equipe favorita? ')
    if resposta=="UFRJ ANALYTICA":
        print('Minha equipe favorita é UFRJ ANALYTICA')
    else:
        print('Frase inesperada')

def pergunta_V2():
    """Pede dois números e imprime o maior, ou, caso
forem iguais imprima que são números iguais"""
    num1= input('Primeiro número:')
    num2= input('Segundo número:')
    if float(num1)==float(num2):
        print('Números iguais')
    else:
        print('O maior número é ', max(float(num1),float(num2)))

#1.3 Estruturas de repetição
def sequencia(numero:int)->list:
    "Recebe um número e imprime cada número de 1 até o número recebido"
    i=1
    lista=[]
    while i<=numero:
        lista.append(i)
        i+=1
    print(lista)

def pede_quant()->int:
    'Pede a quantidade de membros da equipe'
    quant= input('Quantos membros tem a equipe?(responda com um número)')
    return int(quant)

def pergunta_nome(n_membros:int)->str:
    'Pergunta o nome de cada membro e imprime na tela as boas vindas'
    for participante in range(1,n_membros+1):
        nome= input('nome? ')
        print(f'Olá {nome}, seja bem-vindo!')

def main_v3()->str:
    "Função principal"
    n_membros=pede_quant()
    boas_vindas=pergunta_nome(n_membros)
    return boas_vindas

def calculadora(x:float,y:float)->float:
    "Calcula a soma de x e y se eles estiverem entre 0 e 9"
    if x<=9 and x>=0 and y<=9 and y>=0:
        print(f'A soma é {x+y}')
    else:
        print('Escreva números entre 0 e 9')

#1.4 Estruturas de dados
competidores=['João','Carlos','Patrícia','Caio','Letícia']
def nome_na_lista(lista:list)->str:
    'Pede o nome e imprime uma mensagem dizendo se o nome está ou não na lista de competidores'
    nome=input('Nome? ')
    for i in range(len(lista)):
        if lista[i]==nome:
            return 'O nome está na lista'
    else:
        return 'O nome não está na lista'

dicionario={
    'Joao':'Competicao',
    'Carlos':'Desenvolvimento',
    'Patricia':'Competicao',
    'Caio':'Competicao',
    'Leticia':'Gestao'
    }
def mostra_equipe(dicionario:dict)->str:
    'Imprime a equipe que o membro participa'
    nome=input('Nome?')
    for membro in dicionario:
        if membro==nome:
            return dicionario[nome]
    else:
        return 'o nome não existe'

#1.5 Funções e modularização/reaproveitamento de código
def funcao_arg(n:int)->str:
    'recebe um número n e imprime ele n vezes'
    num=str(n)
    saida=num+' '
    return saida*n

def funcao_argV2(numero:int)->str:
    'recebe um número e imprime em cada linha k vezes o número relativo à linha'
    n=0
    while n<numero:
        n+=1
        print(funcao_arg(n))
    return ' '
```
